### Spam Dataset :

```python
sms_reader = pd.read_csv('../input/spam.csv', encoding='latin-1')
sms_reader = sms_reader.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],
                             axis=1)
sms_reader = sms_reader.rename(columns={'v1': 'label', 'v2': 'message'})

sms_reader.head()
```

```python
sms_reader.groupby('label').describe()
```

```python
messages = sms_reader['message']

messages[2]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(analyzer='word', stop_words='english')
message_feature = vectorizer.fit_transform(messages.values.astype('U'))


print(message_feature[2])
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, labels_train, labels_test = train_test_split(
    message_feature, sms_reader['label'])
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

classifier = MultinomialNB()

classifier.fit(features_train, labels_train)

prediction = classifier.predict(features_test)

accuracy = accuracy_score(prediction, labels_test)

accuracy
```
